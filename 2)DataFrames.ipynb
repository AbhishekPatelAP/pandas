#### Creating a DataFrame

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2 = pd.DataFrame(data_list)
```

```python
df2
```

```python
# lets give a column to our list, as there is no column name in our llist
df2 = pd.DataFrame(data_list)
columns = ["Name", 'Age', "City", "Salary"]
df2 = pd.DataFrame(data_list, columns = columns)
df2
```

#### Conditional Selection

```python
df2
```

```python
# i only want to see those people whose age is above 30
df2[df2["Age"] > 30]
```

```python
df2[df2["Salary"] > 65000] # show those whose salary is more than 65000
```

#### AND

```python
# i only want people whose age is above 30 and their city must be paris
df2[(df2["Age"] > 30) & (df2["City"] == "Paris")]
```

```python
# get the people whose age is more than 30 and salary is more than 70000
df2[(df2["Age"] > 30) & (df2["Salary"] > 70000)]
```

#### OR

```python
# i only want people whose age is above 30 or their city can be paris
df2[(df2["Age"] > 30) | (df2["City"] == "Paris")]
```

```python
# get the people whose age is more than 30 or salary is more than 70000
df2[(df2["Age"] > 30) | (df2["Salary"] > 70000)]
```
