**Pivot Tables**

```python
import pandas as pd
import numpy as np

# Create the initial dictionary
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

# Create the DataFrame
df = pd.DataFrame(data)

# Add Month and Quarter columns
df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)

# Display the result
print(df)
```

```python
# will get mean value frim this pivot table, mean is default aggregation with pivot table
pivot1 = pd.pivot_table(df, values= 'Sales', index= 'Region', columns= 'Product')
pivot1
```

```python
# lets find the median from the pivot table
pivot2 = pd.pivot_table(df, values= 'Sales', index= 'Region', columns= 'Product', aggfunc= 'median')
pivot2
```

```python
pivot3 = pd.pivot_table(df, values= ['Sales', 'Units'], index= 'Region', columns= 'Product') # we can do pivot on multiple columns
pivot3
```
