```python
df1
```

```python
def square(x):
    return x ** 2
```

```python
df1['B'] = df1['B'].apply(square) # we can use def function to and change the value of a column as well, now column B has updated values
```

```python
df1
```

```python
df1['D'] = df1['A'].apply(square) # we can add new column as well with name D and put values as square of column A
df1
```
