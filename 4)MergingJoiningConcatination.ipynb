```python
import numpy as np
import pandas as pd
```

#### Merging 2 dataframes

```python
# DataFrame 1: Employees information
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees, salaries) # will get all the common column who has same employee_id(employee_id is taken as default column)
```

```python
pd.merge(employees, salaries, on="employee_id")
```

```python
pd.merge(employees, salaries, on= "employee_id", how= "inner") # similar like inner join in sql, will give the matching data from both dataframe
```

```python
pd.merge(employees, salaries, on= "employee_id", how= "outer") # similar like full outer join in sql, will give the all data from both dataframe
```

```python
pd.merge(employees, salaries, on= "employee_id", how= "left") # similar like left join in sql, will give the all data from left dataframe
```

```python
pd.merge(employees, salaries, on= "employee_id", how= "right") # similar like right join in sql, will give the all data from right dataframe
```

#### Concatenation of 2 dataframes

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
df1
```

```python
df2
```

```python
pd.concat([df1, df2]) # this similar like set(union) in sql where we want to add table one above other. (by default concat works on column)
```

```python
pd.concat([df1, df2], axis= 1) # to concat df row by row we need to give axis there, as axis = 1
```

#### Joining of 2 dataframes

```python
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

# Second DataFrame
df2 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2) # by default its apply left join (join will help to join columns)
```

```python
df1.join(df2, how= 'inner') # will get data only having same index
```

```python
df1.join(df2, how= "outer") # will get all full join table, means both data from both df (like full outer join in sql)
```

```python
df2.join(df1) # left df = df2 and right df = df1
```
