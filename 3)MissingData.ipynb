```python
import numpy as np
import pandas as pd
```

#### Finding Missing Data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
df
```

```python
df.isna() # fird the null value from the df in form of true and false
```

```python
df.isna().sum() # find how many null values are there in every column
```

```python
df.isna().any() # is there any column where there is no any missing values or null value(null values = False)
```

#### Removing Missing Data

```python
df
```

```python
# we can perform removing null operation only on row level
```

```python
df.dropna() # will drop all those row where there is null value in that row
```

```python
df
```

```python
df.dropna(thresh=3) # will get those row who has more than 3 non null values(at least 3 null values should be there, that will be there in our df)
```

```python
df.dropna(thresh=4) # will get those rows who has more than 4 non null values
```

```python
df.dropna(thresh=5)
```

```python
df.dropna(thresh=1)
```

#### Filling the missing Data

```python
df
```

```python
# fill all the missing values with zero # fill works on column wise and drop work on row wise
df.fillna(0) # there will be no any change in our main df, for that we have to use inplace = True
```

```python
df # check main df
```

```python
# fill the all different values for each column(for all null values)
values = {"A": 0, "B": 100, "C": 300, "D": 400}
df.fillna(value= values)
```

```python
df
```

```python
# we can fill with mean or avg value of that column in our df in place of null values (here fill.na(.mean()) will work on column waise)
df.fillna(df.mean())
```
