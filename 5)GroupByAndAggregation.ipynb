```python
import pandas as pd

data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# Group by Category and calculate the sum of Sales
cat = df.groupby('Category')['Sales'].sum() # this df.groupby will convert thois df into object that what we can not see
cat
```

```python
# Group by Store and calculate the sum of Sales
store_s = df.groupby('Store')['Sales'].sum()
store_s
```

```python
# find the average sales by stores
store_m= df.groupby('Store')['Sales'].mean()
store_m
```

```python
# find the maximum sales by Category
ma_cat = df.groupby('Category')['Sales'].max()
ma_cat
```

```python
# find the minimun sales by Category
mi_cat = df.groupby('Category')['Sales'].min()
mi_cat
```

```python
# Group by multiple columns
# Group by Category and Store
cat_store = df.groupby(['Category', 'Store'])['Sales'].sum()
cat_store
```
