```python
import numpy as np
import pandas as pd
```

```python
labels = ['a', 'b', 'c']
my_list = [10, 20, 30]
arr = np.array([10, 20, 30])
d = {1:10, 2:20, 3:30}
```

```python
pd.Series(my_list) # here 0, 1, 2 are default index, but here in pandas we call it as labesl
```

```python
pd.Series(my_list, index=labels) # we can create custom labels as well
```

```python
pd.Series(arr) # we can send one dimensional array to series as well, but not the multi dimensional, because, series itself its one dimensional
```

```python
pd.Series(d) # dictionary has their own keys, so no need of extra indexing
```
